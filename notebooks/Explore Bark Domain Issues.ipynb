####Rename variables

```python
B = Bp
BC = BCp
```

####Matrix BC is not square, so it needs to be full rank in order for us to be able to compute its pseudo-inverse. Let's check its rank.

```python
BC_rank = np.linalg.matrix_rank(BC)

print 'Shape of BC: ', BC.shape
print 'Rank of BC: ', BC_rank
```

Letting k = rank(BC) and BC have dimension m x n, we see that k < m and k < n. So BC is not full rank here, and thus is rank deficient.

####Compute the SVD and the pseudo-inverse of matrix BC using the SVD approach in Numpy

```python
U_BC, S_BC, V_T_BC = np.linalg.svd(BC)

BC_pinv = np.linalg.pinv(BC)
```

```python
# Plot the singular values of BC (decreasing order)
plot(S_BC)
```

```python
# Display the pseudo-inverse of BC
fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(BC_pinv)
ax.set_aspect('auto')
```

```python
# Compute the matrix product of the inverse and the original matrix to see
# how close they are to the identity matrix.
I_BC_papprox = np.dot(BC_pinv, BC)

fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(I_BC_papprox)
ax.set_aspect('auto')
```

Not bad! This is much closer to the identity matrix than what was being produced before using the non-SVD pseudo-inverse approach. But how close is it? One thing we could look at is the square error between the approximate identity matrices and the actual identity matrix.

####Compute the square error between the approximate identity matrices and the actual identity matrix.

```python
diff_BC = I_BC_papprox - np.eye(I_BC_papprox.shape[0])

sq_error_BC = np.mean(np.power(diff_BC,2))

print 'Square Error (BC): ', sq_error_BC
```

The square error from BC is now much smaller! This appears to be working the way we would like for it to.

```python
a = np.dot(BC_pinv, BD)
```

```python
# Form impulse response h(n)
h1 = a[M:0:-1]
h2 = 2*a[0]
#h2 = h2.reshape(h2.shape[0],1)
h3 = a[1:M+1]

h12 = np.concatenate((h1,[h2]))

h = np.concatenate((h12,h3))/2

h = np.divide(h,np.amax(np.abs(h)))
```

###What happens if we compute the pseudo-inverse directly, using the matrix's SVD?

Truncate the already computed SVD for matrix BC, compute the pseudo-inverse explicitly, and look at output of least squares process.

```python
# Truncate the SVD matrices using the rank value of matrix BC
U1 = U_BC[:,0:BC_rank]
S1 = (np.diag(1/S_BC))[0:BC_rank,0:BC_rank]
V1_T = V_T_BC[0:BC_rank,:]
```

```python
# Compute the SVD pseudo-inverse
BC_svd_inv = np.dot(V1_T.T, np.dot(S1,U1.T))
```

```python
# How similar is this pseudo-inverse to the one computed using np.linalg.pinv?
np.allclose(BC_pinv,BC_svd_inv)
```

Apparently there are some noticeable differences for certain components in the matrices.

```python
fig = plt.figure(0, (16,8))

ax = fig.add_subplot(121)
ax.imshow(BC_pinv)
ax.set_aspect('auto')

ax = fig.add_subplot(122)
ax.imshow(BC_svd_inv)
ax.set_aspect('auto')
```

###As a final attempt, let's try one more pseudo-inverse Python function (pinv2).

```python
from scipy.linalg import pinv2

BC_pinv2 = pinv2(BC)
```

```python
fig = figure(0, (16,12))

for i in range(0,20):
    subplot(5,5,i+1)
    plot(BC_pinv2[3*i,:])
    grid(1)
```

We're seeing the same general shape, the oscillations of increasing frequency, high amplitudes, and sudden drop-off to zero towards the end. The main difference is that the amplitudes start out much smaller for the earlier rows compared to the results from using 'pinv' and 'svd' functions.

It appears what's going on is the treatment of the singular values is producing some differences and causing some large amplitudes. Since the SVD method of computing the pseudo-inverse makes use of the inverse of the singular value matrix (this equates to taking one divided by the singular values for the diagonal entries), the amplitudes will be significantly large if the singular values are small. 

To avoid these issues, care must be taken when using the singular values. Ones below a certain threshold should be ignored. These different functions handle this in different ways (i.e. different threshold conditions).
