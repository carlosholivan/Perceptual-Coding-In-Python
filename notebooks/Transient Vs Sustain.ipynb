One of the reasons working with music signals can be challenging is the inherit non-stationarity of musical sounds, especially acoustic instruments. A high level example of this non-stationarity is the inconsistent system behavior between transients and sustains. 

Here we'll make an effort seperate transients and sustains using wavelet decomposition. In practice, the statistical nonstationarity breaks down over more than just transient/sustain behavior, this represents a first attempt at reducing non-stationarity using a transient/sustain paradigm.

##Compute Wavelet Coefficients

```python
import pywt

#Compute SWT
def computeWaveletCoefficients(data, name='db4', numLevels = 2):
    #Create wavelet object
    w = pywt.Wavelet(name)
    
    #Compute wavelet coefficients, use swt to avoid decimation:
    c = pywt.swt(data, w, level=numLevels)
    
    #Put all coefficients into a big matrix, XFF
    XFF = np.zeros((len(data), numLevels + 1))

    XFF[:,0] = c[0][0]
    for i in range(len(c)):
        XFF[:,i+1] = c[i][1]
        
    #Throw out last 1000 samples, wierd stuff tends to happen at the wend of wavelet 
    return XFF[:-1000,:]
```

###Grab HF Wavelet component and square:

```python
#Squared HF Comp.
cHF = cNorm[:,-1]**2
cSquared = cHF/np.max(cHF)

cSquared.shape, amax(cSquared)
```

###Detect transients using adaptive gating method:

```python
import time

def waveletGateAdaptive(input, threshVec, openThresh, samplesToWait, deltaUpParams, deltaDownParams):
    
    startTime = time.clock()
    transientFlags = []
    
    deltaUp = deltaUpParams[0]*sigmoidPrime(linspace(-1*deltaUpParams[1], deltaUpParams[1], 1024))
    deltaDown = deltaDownParams[0]*sigmoidPrime(linspace(-1*deltaDownParams[1],deltaUpParams[1],1024))
    
    gateVal = np.zeros(len(input))
    
    samplesToWaitCount = 0
    
    for i in arange(1, len(input)):
        samplesToWaitCount += 1
        if samplesToWaitCount > samplesToWait:
            if input[i] > threshVec[i]:
                gateVal[i] = gateVal[i-1] + deltaUp[ceil(len(deltaUp)*gateVal[i-1])-1]
                if gateVal[i] > openThresh:
                    transientFlags.append(i)
                    gateVal[i] = 0
                    samplesToWaitCount = 0

            else:
                gateVal[i] = gateVal[i-1] - deltaDown[ceil(len(deltaDown)*gateVal[i-1])-1]

            if gateVal[i] > 1:
                gateVal[i] = 1
            elif gateVal[i] < 0:
                gateVal[i] = 0
                
    print "Time Elapsed = " + str(time.clock()-startTime)
                
    return gateVal, transientFlags
```

Sigmoid is part of gating function:

```python
def generalSigmoid(x, a):
    return 1/(1+exp(-a*x))

def sigmoidPrime(z):
    e = 1/(1+np.exp(z))
    return (1-e)*e
```

```python
g, transientFlags = waveletGateAdaptive(cSquaredShifted, threshVec = 3*threshVec, openThresh = 0.7, \
                        samplesToWait = 600, deltaUpParams = [0.25, 5], deltaDownParams = [0.05, 5])
```

Make gate masking function.

```python
transientDuration = 2048

down = generalSigmoid(linspace(-5,5, transientDuration), -1)

totalLength = len(down)
plot(down)

grid(1)
```

```python
m = np.zeros((cSquaredShifted.shape[0], 2))

for flagCount in range(len(transientFlags)):
    m[transientFlags[flagCount]-128:transientFlags[flagCount], 0] = g[transientFlags[flagCount]-128:transientFlags[flagCount]]
    m[transientFlags[flagCount]:transientFlags[flagCount]+len(down), 0] = down

m[:,1] = 1-m[:,0]
```

Multiply original signals by transient mask

```python
signalKeys = alignedSignals.keys()[0:4]

#Transient and Sustain Dicts:
t = {}
s = {}

for key in signalKeys:
    s[key] = m[:,1]*alignedSignals[key][filterLength:filterLength+m.shape[0]]
    t[key] = m[:,0]*alignedSignals[key][filterLength:filterLength+m.shape[0]]
```

```python
plot(s['piezo'])
plot(t['piezo'])
```

##Make silence-removed version of transients:

```python
tSR = {}

for key in signalKeys:
    tSR[key] = t[key][abs(t[key])>0]
```

```python
plot(tSR['piezo'])
grid(1)
plot(tSR['mic1'])
```
