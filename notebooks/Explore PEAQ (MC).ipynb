##2.4 Calibration of Loudness Scaling Factor
This part honestly confuses the heck out of me right now, I have no idea why it needs to be so complex.

###Explanation:

Some of the perceptual quality factors to come depend on the actual sound pressure level (SPL) of the test signal. However, we don't necessarily, and often times won't, have knowledge of the actual SPL of the test signal. So, a calibration step is needed to fix the mapping from input signal levels to loudness that will occur later.

This calibration step involves the use of a scaling factor which takes into account window and DFT scaling, as well as loudness calibration:

$$
G_{L} = \frac{10^{ \frac{L_{p}}{20} }}{\gamma(f_{c}) \sqrt{\frac{8}{3}} \frac{A_{max}}{4} \frac{N_{F}-1}{N_{F}}}
$$

where $L_{p}$ is the SPL (in dB) corresponding to a full-scale test sinewave signal (in absence of other information, 92dB SPL is used); $\A_{max}$ is the maximum amplitude of the sine (for instance, 32,678 for 16-bit data, or simply 1 for floating point data); $\gamma(f_{c})$ is a factor which varies from 0.84 to 1 depending on where the frequency of the test sine falls relative to the DFT bins.

The $\sqrt{\frac{8}{3}}$ and $\frac{N_{F}-1}{N_{F}}$ terms in the denominator account for scaling from the window applied to the test signal. The $\frac{A_{max}}{4}$ term is to scale by the maximum amplitude/DFT value in the sine test signal.

One possible interpretation of this process is as follows: we desire for the test signal to be properly prepared/scaled to nicely be mapped to the loudness domain. Sound pressure level is defined as follows:

$$
L_{p} = 20 \log_{10} (\frac{p_{rms}}{p_{0}})
$$

To get the pressure value of the signal given the SPL value, you use the following:

$$
P = 10^{\frac{L_{p}}{20}}
$$

The numerator of the calibration factor $G_{L}$ is a mapping from an SPL value back to the pressure domain, which is the domain that the test signal is initially in. When this signal is mapped back to a loudness domain, the numerator value will be inversely processed to produce an SPL value, which we have been able to calibrate ourselves by assuming it to be 92dB SPL. So, we have been able to calibrate the loudness to a specific value, and have removed the effects of the window and DFT scaling factors.

Additionally, we also include one other term in the denominator: the $\gamma(f_{c})$ term. This term is related to the DFT process. The Discrete Fourier Transform (DTFT) is a transform of a discrete (sampled) signal into a (continuous) frequency domain. The Discrete Fourier Transform (DFT) is a sampled version of the DTFT, sampled at equally spaced points along the unit circle. 

<img src="data/PEAQ_data/Kabal_Loudness_Plot.jpg">

Since there is a sampling process involved with the DFT, it is possible to sample DTFT points such that the DFT bins miss the true maximum value in the DTFT. Additionally, the test signal's frequency might not directly map to a DFT bin, which implies a difference in magnitude between it and the nearest bin. The $\gamma(f_{c})$ is used to determine the nearest DFT bin to the test signal's frequency, and to determine the magnitude factor difference between the two. 

```python
def PQ_GL(NF, Amax, fcN, Lp):
    #Scaled Hann window, including loudness scaling
    # Calculate the gain for the Hann Window
    #  - level Lp (SPL) corresponds to a sine with normalized frequency
    #    fcN and a peak value of Amax
    
    W = NF-1
    gp = PQ_gp(fcN, NF, W)
    GL = 10**(Lp/20.) / (gp *Amax/4 *W)
    return GL
    
def PQ_gp(fcN, NF, W):
    # Calculate the peak factor. The signal is a sinusoid windowed with
    # a Hann window. The sinusoid frequency falls between DFT bins. The
    # peak of the frequency response (on a continuous frequency scale) falls
    # between DFT bins. The largest DFT bin value is the peak factor times
    # the peak of the continuous response.
    # fcN - Normalized sinusoid frequency (0-1)
    # NF  - Frame (DFT) length samples
    # NW  - Window length samples

    #Distance to the nearest DFT bin
    df = 1./NF
    k = floor(fcN/df)
    dfN = amin([(k+1)*df - fcN, fcN -k*df])
    
    dfW = dfN*W
    gp = np.sin(np.pi*dfW) / (np.pi*dfW*(1-dfW**2))
    return gp
```

```python
def PQRFFT (x, N, ifn):
    # Calculate the DFT of a real N-point sequence or the inverse
    # DFT corresponding to a real N-point sequence.
    # ifn > 0, forward transform
    #          input x(n)  - N real values
    #          output X(k) - The first N/2+1 points are the real
    #            parts of the transform, the next N/2-1 points
    #            are the imaginary parts of the transform. However
    #            the imaginary part for the first point and the
    #            middle point which are known to be zero are not
    #            stored.
    # ifn < 0, inverse transform
    #          input X(k) - The first N/2+1 points are the real
    #            parts of the transform, the next N/2-1 points
    #            are the imaginary parts of the transform. However
    #            the imaginary part for the first point and the
    #            middle point which are known to be zero are not
    #            stored. 
    #          output x(n) - N real values


    if (ifn > 0):
        X = fft.fft (x, N)
        XR = real(X[0:N/2+1])
        XI = imag(X[1:N/2-1+1])
        X = np.concatenate([XR, XI])
        return X
    else:
        raise Exception('ifft Not Implemented Yet -SW')
```

##Squared Magnitude:

```python
def PQRFFTMSq(X, N):
# Calculate the magnitude squared frequency response from the
# DFT values corresponding to a real signal (assumes N is even)

    X2 = np.zeros(N/2+1)

    X2[0] = X[0]**2
    for k in range(N/2-1):
        X2[k+1] = X[k+1]**2 + X[N/2+k+1]**2

    X2[N/2] = X[N/2]**2
    return X2
```

##Outer and middle Ear Filtering

```python
def PQWOME(f):
    # Generate the weighting for the outer & middle ear filtering
    # Note: The output is a magnitude-squared vector
    
    N = len(f)
    W2 = np.zeros(N)
    
    for k in range(N-1):
        fkHz = float(f[k+1])/1000
        AdB = -2.184 * fkHz**(-0.8) + 6.5 * np.exp(-0.6 * (fkHz - 3.3)**2) - 0.001 * fkHz**(3.6)
        W2[k+1] = 10**(AdB / 10);
    
    return W2
```

For the i'th frequency band, the contribution from the energy in DFT bin K is:

$$
U[i,k] = \frac{max[0, min(f_u[i], \frac{2k+1}{2}\frac{F_s}{N_f})-max(f_l[i],\frac{2k-1}{2}\frac{F_s}{N_f} )]}{\frac{F_s}{N_f}} \tag{13}
$$

This forumala is a bit daunting at first, and the way it's implemented in Kabal confuses the heck out of me. I've opted for a simpler, matrix-based (so hopefully faster), implementation. We can think of this function as a mapping from DFT bins to the non-linear, bark scale frequency bands described above. This gets complicated, becuase at low frequencies, a single DFT bin correspond to  multiple frequency bands, and at high frequencies, a single frequency band will cover multiple DFT bins. Vizualising the 2-D matrix helps.

#2.7 Internal Noise

An offset is added to the band energies to compenstae for internal noise generated in the ear:

$$
E[i] = E_b[i] + E_{IN}[i] 
$$

Internal noise is modeled as:

$$
E_{INdB}(f_{kHz}) = 1.456(f/1000)^{-.08}, \\
E_{IN}[i] = 10^{E_{INdB}(f_c[i])/10}
$$

The energies E[i] are referred to as pitch patterns.

```python
def PQIntNoise (f):
    INdB = 1.456 * (f / 1000.)**(-0.8)
    EIN = 10**(INdB / 10.)
    return EIN
```

##2.8 Frequency Spreading

Lots of formulas in this section!

Probably worth at least considering if these guys can be matrix-ized. 

The spread bark-domain energy response is:

$$
E_s[i] = \frac{1}{B_s[i]}\Big[\sum_{l=0}^{N_c-1}\big(E[l]S(i,l,E[l])\big)^{0.4}\Big]^{1/0.4}
$$

where $N_c$ is the number of filter bands. The normalizing factor is calculated for a reference level of 0 dB for each band,

$$
B_s[i] = \Bigg(\sum_{l=0}^{N_c-1}(S(i, l, E_0))^{0.4} \Bigg)^{\frac{1}{0.4}}
$$

where $E_0$ = 1 (0 dB).

The spreading function is: 

$$
S(i, l, E) = \frac{1}{A(l,E)}10^{S_{dB}(i, l, E)/10}
$$

where the normalizing term is chosen to give a unit area for each center frequency l. The spreading function in dB is traingular. 

$$
   S_{dB}(i, l , E) = \left\{
     \begin{array}{lr}
       27(i-l)\Delta z & , i \leq l,\\
       \big[-24-\frac{230}{f_c[l]}+2log_{10}(E)\big](i-l)\Delta z & , i \geq l.
     \end{array}
   \right.
$$

I want to develop a little more intuitive on this equation. It's a funciton of 3 varialbes, which is a little confusing. It tells us how much the energy is other bins affects a certain bin.I think it's the effect of bins l on bin i. 

|  **Variable** | **Description**  |
|:---:|:---:|
|  $i$ | Bin being affected |

|  $E$ | Energy in other bin $l$ |
|  $S(i, l, E)$ | Function that tells us how much energy from bin $l$ ends up in bin $i$, also depends on energy in bin $l$ |

##Spread Energy Computations - Modified with Regrouping and Simplifications

```python
def PQ_SpreadCB(E, Bs):
    
    # Get the frequency arrays and other parameter
    # values
    (Nc, fc, fl, fu, dz) = PQCB()
    
    e = 0.4 # Commonly used power value
    
    # Initialize arrays for storage. These values are used
    # in each iteration (summed over, multiplied, raised to
    # powers, etc.) when computing the spread Bark-domain
    # energy Es.
    #
    # aUCEe is for the product of bin-dependent (index l)
    # term aC, energy-dependent (E) term aE, and
    # term aU.
    #
    # Ene is (E[l]/A(l,E[l]))^e, stored for each index l
    #
    # Es is the overall spread Bark-domain energy
    #
    aUCEe = np.zeros(Nc)
    Ene = np.zeros(Nc)
    Es = np.zeros(Nc)
    
    # Calculate energy-dependent terms
    aL = 10**(2.7*dz)
    for l in xrange(Nc):
        aUC = 10**((-2.4 - 23/fc[l])*dz)
        aUCE = aUC * (E[l]**(0.2*dz))
        gIL = (1 - aL**(-1*(l+1))) / (1 - aL**(-1))
        gIU = (1 - (aUCE)**(Nc-l)) / (1 - aUCE)
        En = E[l] / (gIL + gIU - 1)
        aUCEe[l] = aUCE**(e)
        Ene[l] = En**(e)
    
    # Lower spreading
    Es[Nc-1] = Ene[Nc-1]
    aLe = aL**(-1*e)
    for i in xrange((Nc-2),-1,-1):
        Es[i] = aLe*Es[i+1] + Ene[i]
    
    
    # Upper spreading (i > m)
    for i in xrange(0,(Nc-1)):
        r = Ene[i]
        a = aUCEe[i]
        for l in xrange((i+1),Nc):
            r = r*a
            Es[l] = Es[l] + r
            
    # Normalize the values by the normalization factor
    for i in xrange(0,Nc):
        Es[i] = (Es[i]**(1/e)) / Bs[i]
        
    return Es
```

###Time-domain spreading

```python
def PQ_timeSpread(Es, Ef):
    
    # Define/obtain parameters
    (Nc, fc, fl, fu, dz) = PQCB()
    Fs = 48000
    NF = 2048
    Nadv = NF/2
    Fss = float(Fs)/Nadv
    tau_100 = 0.030
    tau_min = 0.008
    alpha = PQtConst(tau_100, tau_min, fc, Fss)
    
    # Allocate storage
    Ehs = np.zeros(Nc)
    
    # Time domain smoothing
    for i in xrange(Nc):
        Ef[i] = alpha[i]*Ef[i] + (1-alpha[i])*Es[i]
        Ehs[i] = max(Ef[i],Es[i])
        
    #plot(Ehs)
    return Ehs, Ef
```

```python
def PQDFTFrame(x):
    NF = 2048 # Frame size (samples)

    Amax = 32768
    #Amax = 1
    fc = 1019.5
    Fs = 48000
    Lp = 92
    
    # Set up the window (including all gains)
    GL = PQ_GL(NF, Amax, fc/Fs, Lp)
    hw = GL * PQHannWin(NF)

    # Window the data
    xw = hw * x

    # DFT (output is real followed by imaginary)
    X = PQRFFT(xw, NF, 1)

    # Squared magnitude
    X2 = PQRFFTMSq(X, NF)
    
    return X2
```

```python
def PQgroupCB(X2):
    # Group a DFT energy vector into critical bands
    # X2 - Squared-magnitude vector (DFT bins)
    # Eb - Excitation vector (fractional critical bands)

    Emin = 1e-12

    # Set up the DFT bin to critical band mapping
    NF = 2048
    Fs = 48000
    Nc, fc, fl, fu, dz = PQCB()
    
    df = float(Fs) / NF
    
    U = np.zeros((NF/2+1, Nc))

    for k in range(NF/2+1):
        for i in range(Nc):
            temp = (np.amin([fu[i], (k+0.5)*df]) - np.amax([fl[i], (k-0.5)*df])) / df
            U[k, i] = np.amax([0, temp])
            
    Eb = np.zeros(Nc)

    Eb = np.dot(X2,U)
    Eb[Eb<Emin] = Emin
    
    return Eb

    # Allocate storage
    #Eb = zeros (1, Nc);

    # Compute the excitation in each band
    #for (i = 0:Nc-1)
    #    Ea = Ul(i+1) * X2(kl(i+1)+1);       % First bin
    #    for (k = (kl(i+1)+1):(ku(i+1)-1))
    #        Ea = Ea + X2(k+1);              % Middle bins
    #    end
    #    Ea = Ea + Uu(i+1) * X2(ku(i+1)+1);  % Last bin
    #    Eb(i+1) = max(Ea, Emin);
    #end
```

```python
def PQspreadCB(E):
    # Spread an excitation vector (pitch pattern) - FFT model
    # Both E and Es are powers

    Nc = len(E)
    Bs = PQ_SpreadCB(np.ones(Nc), np.ones(Nc))

    Es = PQ_SpreadCB(E, Bs)
    
    return Es
```

```python
def PQ_excitCB(X2):

    NF = 2048
    Fs = 48000
    f = linspace (0, Fs/2, NF/2+1)
    W2 = PQWOME (f)
    Nc, fc, fl, fu, dz = PQCB()
    EIN = PQIntNoise(fc)
    
    # Allocate storage
    Xw2 = np.zeros((2, NF/2+1))

    # Outer and middle ear filtering
    Xw2[0,:] = W2 * X2[0,0:NF/2+1]
    Xw2[1,:] = W2 * X2[1,0:NF/2+1]

    XwN2 = np.zeros(NF/2+1)

    # Form the difference magnitude signal
    for k in range(NF/2):
        XwN2[k] = (Xw2[0,k] - 2 * np.sqrt (Xw2[0,k] * Xw2[1,k]) \
               + Xw2[1,k])
    
    # Group into partial critical bands
    Eb = np.zeros((2,Nc))
    
    Eb[0,:] = PQgroupCB(Xw2[0,:])
    Eb[1,:] = PQgroupCB(Xw2[1,:])
    EbN     = PQgroupCB(XwN2)

    # Add the internal noise term => "Pitch patterns"
    E = np.zeros((2,Nc))
    
    E[0,:] = Eb[0,:] + EIN
    E[1,:] = Eb[1,:] + EIN

    # Critical band spreading => "Unsmeared excitation patterns"
    #Es(1,:) = PQSpreadCB (E(1,:), Version);
    #Es(2,:) = PQSpreadCB (E(2,:), Version);
    Es = np.zeros((2,Nc))
    
    Es[0,:] = PQspreadCB(E[0,:])
    Es[1,:] = PQspreadCB(E[1,:])
    
    return EbN, Es
```

```python
def PQadapt(Ehs, Fmem):
    # Level and pattern adaptation
    # For now, treat Fmem as a dictionary with the following key/value pairs
    # necessary for this function:
    #   Fmem['P'] - (2 x Nc) array
    #   Fmem['Rn'] - (1 x Nc) array
    #   Fmem['Rd'] - (1 x Nc) array
    #   Fmem['PC'] - (2 x Nc) array

    Nc, fc, fl, fu, dz = PQCB()
    NF = 2048
    Nadv = NF / 2
    
    #Model = Mod #???
    Fs = 48000
    Fss = Fs / Nadv
    tau_100 = 0.050
    tau_min = 0.008
    alpha = PQtConst(tau_100, tau_min, fc, Fss)
    #M1, M2 = PQ_M1M2(Model) # Should only be FFT model, I believe
    M1, M2 = PQ_M1M2()

    # Allocate memory
    EP = np.zeros((2, Nc))
    R = np.zeros((2, Nc))

    # Smooth the excitation patterns
    # Calculate the correlation terms
    sn = 0
    sd = 0
    for k in xrange(Nc):
        # Fmem.P ???
        #Fmem.P(1,m+1) = a(m+1) * Fmem.P(1,m+1) + b(m+1) * Ehs(1,m+1)
        #Fmem.P(2,m+1) = a(m+1) * Fmem.P(2,m+1) + b(m+1) * Ehs(2,m+1)
        #sn = sn + sqrt (Fmem.P(2,m+1) * Fmem.P(1,m+1))
        #sd = sd + Fmem.P(2,m+1)
        Fmem['P'][0,k] = alpha[k]*Fmem['P'][0,k] + (1-alpha[k])*Ehs[0,k]
        Fmem['P'][1,k] = alpha[k]*Fmem['P'][1,k] + (1-alpha[k])*Ehs[1,k]
        sn = sn  + np.sqrt(Fmem['P'][1,k] * Fmem['P'][0,k])
        sd = sd + Fmem['P'][1,k]

    # Level correlation (momentary correction factor)
    CL = (sn / sd)**2

    # Level correction of excitation patterns
    for m in xrange(Nc):
        # Scale one of the signals to match levels
        if (CL > 1):
            EP[0,m] = Ehs[0,m] / CL
            EP[1,m] = Ehs[1,m]
        else:
            EP[0,m] = Ehs[0,m]
            EP[1,m] = Ehs[1,m] * CL

        # Calculate a pattern match correction factor
        #Fmem.Rn(m+1) = a(m+1) * Fmem.Rn(m+1) + EP(2,m+1) * EP(1,m+1);
        #Fmem.Rd(m+1) = a(m+1) * Fmem.Rd(m+1) + EP(1,m+1) * EP(1,m+1);
        Fmem['Rn'][m] = alpha[m]*Fmem['Rn'][m] + EP[1,m]*EP[0,m]
        Fmem['Rd'][m] = alpha[m]*Fmem['Rd'][m] + EP[0,m]*EP[0,m]
        if (Fmem['Rd'][m] <= 0) or (Fmem['Rn'][m] <= 0):
            #error ('>>> PQadap: Rd or Rn is zero');
            print 'Error: PQadap: Rd or Rn is zero'
        
        if (Fmem['Rn'][m] >= Fmem['Rd'][m]):
            R[0,m] = 1
            R[1,m] = Fmem['Rd'][m] / Fmem['Rn'][m]
            #R(1,m+1) = 1;
            #R(2,m+1) = Fmem.Rd(m+1) / Fmem.Rn(m+1);
        else:
            R[0,m] = Fmem['Rn'][m] / Fmem['Rd'][m]
            R[1,m] = 1
            #R(1,m+1) = Fmem.Rn(m+1) / Fmem.Rd(m+1);
            #R(2,m+1) = 1;

    # Average the correction factors over M channels and smooth with time
    # Create spectrally adapted patterns
    for m in xrange(Nc):
        iL = max(m - M1, 0)
        iU = min(m + M2, Nc-1)
        s1 = 0
        s2 = 0
        
        for i in xrange(iL,iU+1):
            s1 = s1 + R[0,i]
            s2 = s2 + R[1,i]
        
        #Fmem.PC(1,m+1) = a(m+1) * Fmem.PC(1,m+1) + b(m+1) * s1 / (iU-iL+1);
        #Fmem.PC(2,m+1) = a(m+1) * Fmem.PC(2,m+1) + b(m+1) * s2 / (iU-iL+1);
        Fmem['PC'][0,m] = alpha[m]*Fmem['PC'][0,m] + (1-alpha[m]) * s1 / (iU-iL+1)
        Fmem['PC'][1,m] = alpha[m]*Fmem['PC'][1,m] + (1-alpha[m]) * s2 / (iU-iL+1)

        # Final correction factor => spectrally adapted patterns
        #EP(1,m+1) = EP(1,m+1) * Fmem.PC(1,m+1);
        #EP(2,m+1) = EP(2,m+1) * Fmem.PC(2,m+1);
        EP[0,m] = EP[0,m] * Fmem['PC'][0,m]
        EP[1,m] = EP[1,m] * Fmem['PC'][1,m]

    return EP, Fmem
```

```python
def PQmovBW(X2):
    # Bandwidth tests
    
    # fx and fl will always be the same values
    NF = 2048
    Fs = 48000
    fx = 21586
    kx = int(round(NF * float(fx)/Fs)) # 921
    fl = 8109
    kl = int(round(NF * float(fl)/Fs)) # 346
    FRdB = 10 # Ref. signal to exceed threshold level by 10dB
    FR = 10**(FRdB/10)
    FTdB = 5 # Test signal to exceed threshold level by 5dB
    FT = 10**(FTdB/10)
    N = NF/2
    
    BW = dict()
    
    # This is the method Kabal uses to find the threshold level.
    # The for loop seems unneccessary and not as efficient as
    # simply using a max operation on an array of the subset of DFT square mag.
    # values above 21.6kHz.
    
    #Xth = X2[1,kx]
    #for k in xrange(kx,N-1):
    #    Xth = max(Xth, X2[1,k+1])
        
    # Using the max operation on the subset of array values.
    # When tested, was about 0.1ms faster, produces same result,
    # and is cleaner, so I'll stick with this version.
    Xth = max(X2[1,kx:N])
    
    # BWRef and BWTest remain negative if the BW of the test signal
    # does not exceed FR * Xth for kx-1 <= k <= kl+1
    BW['BWRef'] = -1
    BWRef = -1
    XthR = FR * Xth # Reference signal threshold level
    for k in xrange(kx-1,kl,-1):
        if (X2[0,k+1] >= XthR):
            BWRef = k+1
            break
            
    BW['BWRef'] = BWRef
    
    BW['BWTest'] = -1
    BWTest = -1
    XthT = FT * Xth # Test signal threshold level
    for k in xrange(BWRef-1,-2,-1):
        if (X2[1,k+1] >= XthT):
            BWTest = k+1
            break
            
    BW['BWTest'] = BWTest
    
    return BW
```

##Model Output Variable - Total NMR (Noise-to-Mask Ratio)

```python
def PQ_MaskOffset(dz, Nc):
    gm = np.zeros(Nc)
    for k in xrange(Nc):
        if (k <= 12./dz):
            mdB = 3
        else:
            mdB = 0.25*k*dz  
        gm[k] = 10**(-1*float(mdB)/10) 
    return gm
```

```python
def PQmovNMRB(EbN, Ehs):
    # Noise-to-mask ratio
    # NMR['NMRavg'] = average NMR
    # NMR['NMRmax'] = max. NMR
    
    NMR = dict()
    
    # Get ___
    Nc, fc, fl, fu, dz = PQCB()
    gm = PQ_MaskOffset(dz, Nc)
    
    NMRmax = 0
    NMRm = 0
    s = 0
    # Don't need to store the values in an array,
    # since in the end we mainly care about the average
    # and maximum values. But for debugging and validation
    # purposes, I'll include this.
    R_NM = np.zeros(Nc)
    
    for k in xrange(Nc):
        NMRm = EbN[k] / (gm[k] * Ehs[k])
        R_NM[k] = NMRm # Remove later!
        s = s + NMRm
        
        if (NMRm > NMRmax):
            NMRmax = NMRm
            
    NMR['NMRmax'] = NMRmax
    NMR['NMRavg'] = float(s)/Nc
    
    return NMR
```
