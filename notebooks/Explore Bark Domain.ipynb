###About PEAQ
PEAQ is a standardized algorithm used for objective measurements of perceived audio quality, and is based on generally accepted psychoacoustic principles. The method incorporates models of both the peripheral ear component as well as a cognitive component to account for the processes associated with quality judgments.

Given a processed signal and the time-aligned original signal, concurrent frames of the signals are mapped to a basilar membrane representation, and differences in these transformed frames are analyzed in both time and frequency domain by a cognitive model. The overall goal of this analysis is to obtain a quality measure similar to a Subjective Difference Grade (SDG) acquired from listening tests with human participants. This output is called the Objective Difference Grade (ODG).

The processing steps extract perceptual features, which are used for predicting the quality measure. These intermediate features, called Model Output Variables (MOVs), provide potential insight into different psychoacoustic dimensions. The MOVs are combined together to produce the final score. There are two versions of PEAQ (Basic and Advanced), which make use of different number and type of MOV parameters. The Basic uses 11, whereas the Advanced uses only 5 [1].

Since PEAQ is a standard, the algorithm is licensed and targeted more at telecommunication companies. However, various implementation attempts are floating around the Internet. One commonly mentioned one is by Peter Kabal of the Multimedia Signal Processing lab at McGill University. Kabal did an extensive examination and interpretation of PEAQ, which resulted in a report titled "An Examination and Interpretation of ITU-R BS.1387: Perceptual Evaluation of Audio Quality" and a MATLAB implementation library called PQevalAudio [2]. This code can be found floating around on the Internet (we acquired it through a GitHub account for a CCRMA student).

## 2. Time to Frequency Domain

### 2.1 Input

Peak assumes a sampling frequency of $F_s$ = 48,000 and a frame length of $N_F$ = 2048. Frames are advances in steps of 1024 samples.

### 2.2 Windowing

Windowing is done with a Hann Window:

$$
   h[n, N_f] = \left\{
     \begin{array}{lr}
       \frac{1}{2}\big[1-cos(\frac{2\pi n}{N_f-1})\big] & , 0 \leq N_f-1,\\
       \ 0, & , otherwise.
     \end{array}
   \right.\tag{1}
$$

```python
#Method to make hanning window, given lenth of window:
def PQHannWin(NF):
    n = np.arange(0, NF)
    hw = 0.5*(1-np.cos(2*np.pi*n/(NF-1)))
    return hw
```

```python
#And plot
hw = PQHannWin(2048)
plot(hw)
grid(1)
```

The sum of the squared window is $\frac{3}{8}(N_f-1)$ should be used for scaling. However, the standard uses: $\sqrt{\frac{3}{8}}$.

### 2.3 Frequency Domain Coefficients

Conversion to the frequency domain is accomplished with a scaled DFT.

$$
X[k] = \frac{1}{N_F}\sum^{N_F-1}_{n=0}h_w[n]x[n]e^{-j2\pi k/N_F} \tag{4}
$$

Again, the scalng in the standard is strange.

```python
#Using the same function names as Kabal:

def PQRFFT (x, N, ifn):
    # Calculate the DFT of a real N-point sequence or the inverse
    # DFT corresponding to a real N-point sequence.
    # ifn > 0, forward transform
    #          input x(n)  - N real values
    #          output X(k) - The first N/2+1 points are the real
    #            parts of the transform, the next N/2-1 points
    #            are the imaginary parts of the transform. However
    #            the imaginary part for the first point and the
    #            middle point which are known to be zero are not
    #            stored.
    # ifn < 0, inverse transform
    #          input X(k) - The first N/2+1 points are the real
    #            parts of the transform, the next N/2-1 points
    #            are the imaginary parts of the transform. However
    #            the imaginary part for the first point and the
    #            middle point which are known to be zero are not
    #            stored. 
    #          output x(n) - N real values


    if (ifn > 0):
        X = fft.fft (x, N)
        XR = real(X[0:N/2+1])
        XI = imag(X[1:N/2-1+1])
        X = np.concatenate([XR, XI])
        return X
    else:
        raise Exception('ifft Not Implemented Yet -SW')
```

```python
def PQRFFTMSq(X, N):
# Calculate the magnitude squared frequency response from the
# DFT values corresponding to a real signal (assumes N is even)

    X2 = np.zeros(N/2+1)

    X2[0] = X[0]**2
    for k in range(N/2-1):
        X2[k+1] = X[k+1]**2 + X[N/2+k+1]**2

    X2[N/2] = X[N/2]**2
    return X2
```

### 2.4 Calibration of the Loudness Scaling Factor

Some of the perceptual quality factors to come depend on the actual sound pressure level (SPL) of the test signal. However, we don't necessarily, and often times won't, have knowledge of the actual SPL of the test signal. So, a calibration step is needed to fix the mapping from input signal levels to loudness that will occur later.

This calibration step involves the use of a scaling factor which takes into account window and DFT scaling, as well as loudness calibration:

$$
G_{L} = \frac{10^{ \frac{L_{p}}{20} }}{\gamma(f_{c}) \sqrt{\frac{8}{3}} \frac{A_{max}}{4} \frac{N_{F}-1}{N_{F}}} \tag{5}
$$


where $L_{p}$ is the SPL (in dB) corresponds to a full-scale test sinewave signal (in absence of other information, 92dB SPL is used); $A_{max}$ is the maximum amplitude of the sine (for instance, 32,678 for 16-bit data, or simply 1 for floating point data); $\gamma(f_{c})$ is a factor which varies from 0.84 to 1 depending on where the frequency of the test sine falls relative to the DFT bins.

The $\sqrt{\frac{8}{3}}$ and $\frac{N_{F}-1}{N_{F}}$ terms in the denominator account for scaling from the window applied to the test signal. The $\frac{A_{max}}{4}$ term is to scale by the maximum amplitude/DFT value in the sine test signal.


One possible interpretation of this process is as follows: we desire for the test signal to be properly prepared/scaled to nicely be mapped to the loudness domain. Sound pressure level is defined as follows:

$$
L_{p} = 20 \log_{10} (\frac{p_{rms}}{p_{0}})
$$

To get the pressure value of the signal given the SPL value, you use the following:

$$
P = 10^{\frac{L_{p}}{20}}
$$

The numerator of the calibration factor $G_{L}$ is a mapping from an SPL value back to the pressure domain, which is the domain that the test signal is initially in. When this signal is mapped back to a loudness domain, the numerator value will be inversely processed to produce an SPL value, which we have been able to calibrate ourselves by assuming it to be 92dB SPL. So, we have been able to calibrate the loudness to a specific value, and have removed the effects of the window and DFT scaling factors.

Additionally, we also include one other term in the denominator: the $\gamma(f_{c})$ term. This term is related to the DFT process. The Discrete Fourier Transform (DTFT) is a transform of a discrete (sampled) signal into a (continuous) frequency domain. The Discrete Fourier Transform (DFT) is a sampled version of the DTFT, sampled at equally spaced points along the unit circle. 

<img src="data/PEAQ_data/Kabal_Loudness_Plot.png">



Since there is a sampling process involved with the DFT, it is possible to sample DTFT points such that the DFT bins miss the true maximum value in the DTFT. Additionally, the test signal's frequency might not directly map to a DFT bin, which implies a difference in magnitude between it and the nearest bin. The $\gamma(f_{c})$ is used to determine the nearest DFT bin to the test signal's frequency, and to determine the magnitude factor difference between the two. 

```python
def PQ_GL(NF, Amax, fcN, Lp):
    #Scaled Hann window, including loudness scaling
    # Calculate the gain for the Hann Window
    #  - level Lp (SPL) corresponds to a sine with normalized frequency
    #    fcN and a peak value of Amax
    
    W = NF-1
    gp = PQ_gp(fcN, NF, W)
    GL = 10**(Lp/20.) / (gp *Amax/4 *W)
    return GL
    
def PQ_gp(fcN, NF, W):
    # Calculate the peak factor. The signal is a sinusoid windowed with
    # a Hann window. The sinusoid frequency falls between DFT bins. The
    # peak of the frequency response (on a continuous frequency scale) falls
    # between DFT bins. The largest DFT bin value is the peak factor times
    # the peak of the continuous response.
    # fcN - Normalized sinusoid frequency (0-1)
    # NF  - Frame (DFT) length samples
    # NW  - Window length samples

    #Distance to the nearest DFT bin
    df = 1./NF
    k = floor(fcN/df)
    dfN = amin([(k+1)*df - fcN, fcN -k*df])
    
    dfW = dfN*W
    gp = np.sin(np.pi*dfW) / (np.pi*dfW*(1-dfW**2))
    return gp
```

### 2.5 Outer and Middle Ear Modelling

Frequency response of the outer and middle ear. 

```python
def PQWOME(f):
    # Generate the weighting for the outer & middle ear filtering
    # Note: The output is a magnitude-squared vector
    
    N = len(f)
    W2 = np.zeros(N)
    
    for k in range(N-1):
        fkHz = float(f[k+1])/1000
        AdB = -2.184 * fkHz**(-0.8) + 6.5 * np.exp(-0.6 * (fkHz - 3.3)**2) - 0.001 * fkHz**(3.6)
        W2[k+1] = 10**(AdB / 10);
    
    return W2
```

```python
f = np.linspace(0, Fs/2, NF/2+1)
W2 = PQWOME (f);
```

####Grouping Frequency Bins:

For the i'th frequency band, the contribution from the energy in DFT bin K is:

$$
U[i,k] = \frac{max[0, min(f_u[i], \frac{2k+1}{2}\frac{F_s}{N_f})-max(f_l[i],\frac{2k-1}{2}\frac{F_s}{N_f} )]}{\frac{F_s}{N_f}} \tag{13}
$$

We can think of this function as a mapping from DFT bins to the non-linear, bark scale frequency bands described above. This gets complicated, becuase at low frequencies, a single DFT bin correspond to  multiple frequency bands, and at high frequencies, a single frequency band will cover multiple DFT bins. Vizualising the 2-D matrix helps.

```python
(Nc, fc, fl, fu, dz) = PQCB()
```

```python
U = np.zeros((NF/2+1, Nc))
df = float(Fs) / NF;

for k in range(NF/2+1):
    for i in range(Nc):
        temp = (np.amin([fu[i], (k+0.5)*df]) - np.amax([fl[i], (k-0.5)*df])) / df
        U[k, i] = np.amax([0, temp])
```

```python
fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(flipud(U.T))
ax.set_aspect('auto')
xlabel('DFT Bin')
ylabel('Frequency Band')
```

So here we see the mapping. The matrix is mostly zeros, which makes sense, as a given frequency band will only draw from a few DFT bins at most. We can think of the U values as "blending" multiple BINs, if a single DFT bin corresponds to a single frequency band, then U is one.

The energy inband i for a DFT-based signal is:

$$
E_a[i] = \sum_{k=k_l[i]}^{k_u[i]}U[i,k]\vert{X_w[k]}\vert^2 \tag{14}
$$

$$
E_b[i] = max(E_a[i], E_min) \tag{16}

```python
def fft2barkmx(nfft, fs, nfilts=0, width=1.0, minfreq=0, maxfreq=24000):

    #function wts = fft2barkmx(nfft, sr, nfilts, width, minfreq, maxfreq)
    # wts = fft2barkmx(nfft, sr, nfilts, width, minfreq, maxfreq)
    #      Generate a matrix of weights to combine FFT bins into Bark
    #      bins.  nfft defines the source FFT size at sampling rate sr.
    #      Optional nfilts specifies the number of output bands required 
    #      (else one per bark), and width is the constant width of each 
    #      band in Bark (default 1).
    #      While wts has nfft columns, the second half are all zero. 
    #      Hence, Bark spectrum is fft2barkmx(nfft,sr)*abs(fft(xincols,nfft));
    # 2004-09-05  dpwe@ee.columbia.edu  based on rastamat/audspec.m
    
    maxfreq = fs/2

    min_bark = hz2bark(minfreq)
    nyqbark = hz2bark(maxfreq) - min_bark
    if (nfilts == 0):
      nfilts = np.ceil(nyqbark)+1

    wts = np.zeros((nfilts, nfft))

    # bark per filt
    step_barks = nyqbark/(nfilts-1)

    # Frequency of each FFT bin in Bark
    #binbarks = hz2bark(xrange((nfft/2)+1)*fs/nfft)
    binbarks = hz2bark(np.linspace(0,(nfft/2),(nfft/2)+1)*fs/nfft)
    #binbarks = hz2bark([0:nfft/2]*sr/nfft)

    for i in xrange(nfilts):
      f_bark_mid = min_bark + (i) * step_barks
      # Linear slopes in log-space (i.e. dB) intersect to trapezoidal window
      lof = np.add(binbarks, (-1*f_bark_mid - 0.5))
      hif = np.add(binbarks, (-1*f_bark_mid + 0.5))
      #lof = (binbarks - f_bark_mid - 0.5)
      #hif = (binbarks - f_bark_mid + 0.5)
      #wts(i,1:(nfft/2+1)) = 10.^(min(0, min([hif; -2.5*lof])/width))
      wts[i,0:(nfft/2)+1] = 10**(np.minimum(0, np.minimum(np.divide(hif,width), np.multiply(lof,-2.5/width))))
    
    return wts

def hz2bark(f):
    #       HZ2BARK         Converts frequencies Hertz (Hz) to Bark
    # taken from rastamat
    z = 6 * np.arcsinh(f/600)
    return z
```

##Bark Domain Transform Class

```python
def hz2bark(f):
    #       HZ2BARK         Converts frequencies Hertz (Hz) to Bark
    # taken from rastamat
    z = 6 * np.arcsinh(f/600)
    return z
```

```python
import numpy as np

class Bark:
    '''
    
    '''
    def __init__(self, nfft=2048, nfilts=109, fs=48000, width=1.0, minfreq=0, maxfreq=24000):
        self.nfft = nfft
        self.nfilts = nfilts
        self.fs = fs
        self.width = width
        self.min_freq = minfreq
        self.max_freq = maxfreq
        self.nfreqs = nfft/2
        
        # Compute the forward Bark transform matrix
        # (Fourier (FFT) frequency domain to Bark domain)
        self.W = self.fft2barkmx()
        
        # Compute the backwards Bark transform matrix
        # (Bark domain back to Fourier (FFT) domain)
        self.W_inv = self.bark2fftmx()
        
        print "Bark transform weight matrix precomputed, ready to go!"
        
        
    def fft2barkmx(self):

        #function wts = fft2barkmx(nfft, sr, nfilts, width, minfreq, maxfreq)
        # wts = fft2barkmx(nfft, sr, nfilts, width, minfreq, maxfreq)
        #      Generate a matrix of weights to combine FFT bins into Bark
        #      bins.  nfft defines the source FFT size at sampling rate sr.
        #      Optional nfilts specifies the number of output bands required 
        #      (else one per bark), and width is the constant width of each 
        #      band in Bark (default 1).
        #      While wts has nfft columns, the second half are all zero. 
        #      Hence, Bark spectrum is fft2barkmx(nfft,sr)*abs(fft(xincols,nfft));
        # 2004-09-05  dpwe@ee.columbia.edu  based on rastamat/audspec.m
        
        minfreq = self.min_freq
        maxfreq = self.max_freq

        min_bark = hz2bark(minfreq)
        nyqbark = hz2bark(maxfreq) - min_bark
        
        width = self.width
        fs = self.fs
        nfft = self.nfft
        nfilts = self.nfilts
        if (nfilts == 0):
          nfilts = np.ceil(nyqbark)+1

        wts = np.zeros((nfilts, nfft))

        # Bark per filt
        step_barks = nyqbark/(nfilts-1)

        # Frequency of each FFT bin in Bark
        binbarks = hz2bark(np.linspace(0,(nfft/2),(nfft/2)+1)*fs/nfft)

        for i in xrange(nfilts):
          f_bark_mid = min_bark + (i)*step_barks
          # Linear slopes in log-space (i.e. dB) intersect to trapezoidal window
          lof = np.add(binbarks, (-1*f_bark_mid - 0.5))
          hif = np.add(binbarks, (-1*f_bark_mid + 0.5))
          wts[i,0:(nfft/2)+1] = 10**(np.minimum(0, \
                        np.minimum(np.divide(hif,width), np.multiply(lof,-2.5/width))))
        
        return wts

    
    def bark2fftmx(self):
        
        W = self.W
        W_short = W[:, 0:self.nfreqs]
        
        # Need to transpose and fix up the forward weight matrix
        # to get the backwards weight matrix
        WW = np.dot(W_short.T,W_short)

        WW_diag_max = np.maximum(mean(diag(WW))/100, sum(WW,1))
        WW_diag_max = np.reshape(WW_diag_max,(WW_diag_max.shape[0],1))
        W_inv_denom = tile(WW_diag_max,(1,self.nfilts))

        W_inv = np.divide(W_short.T, W_inv_denom)

        return W_inv
    
    def forward(self, spectrum):
        
        W_short = self.W[:, 0:self.nfreqs]
        bark_spectrum = np.dot(W_short,spectrum)
        
        return bark_spectrum
        
        
    def backward(self, bark_spectrum):
        
        spectrum = np.dot(self.W_inv,bark_spectrum)
        
        return spectrum
```
