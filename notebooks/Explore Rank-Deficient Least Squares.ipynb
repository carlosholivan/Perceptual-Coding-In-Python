In the process of exploring Bark domain filter design, the main issue that I was confronted with was a rank-deficient matrix in the least squares problem.

When considering the frequency-domain least squares approach, when applying the Bark transform to the problem formulation $\textbf{BCa} \approx \textbf{BD}$, the analytic solution can be formulated as:

$$\textbf{a} = ((BC)^{T}BC)^{-1} (BC)^{T} \textbf{BD}$$

This assumes that matrix $BC$ is "invertible," i.e. the Moore-Penrose pseudoinverse of BC exists and can be computed. However, the rank deficiency of $BC$ makes this not work in the end.

Before exploring more into the problem involving the Bark domain filter design problem, I will first explore matrix inversion in some detail.

##Matrix Inversion

Let $\textbf{A}$ be a square **n x n** matrix. If $\textbf{A}$ is invertible, then $\textbf{A}$ has full rank, i.e. rank($\textbf{A}$) = n (the columns of $\textbf{A}$ are linearly independent).

Now, consider a non-square matrix $\textbf{A}$ that is **m x n**. Non-square matrices don't have inverses. However, the Moore-Penrose pseudo-inverse can be defined for an **m x n** matrix (m and n not equal). This pseudo-inverse of $\textbf{A}$ is defined as an **n x m** matrix $A^{+}$.

Matrix $A^{+}$ exists for any matrix A, but when the latter has full rank, $A^{+}$ can be expressed as a simple algebraic formula.
* When A has full column rank (matrix $A^{T}A$ is invertible), then $A^{+} = (A^{T}A)^{-1}A^{T}$ (left inverse)
* When A has full row rank (matrix $AA^{T}$ is invertible), then $A^{+} = A^{T}(AA^{T})^{-1}$ (right inverse)

However, when A is not full rank, then these formulas cannot be used. More generally, the pseudo-inverse is best computed using the Singular Value Decomposition (SVD).

###Singular Value Decomposition (SVD)

Let $\textbf{A}$ be an **m x n** real matrix. Then there exists orthogonal matrices $\textbf{U}$ (**m x m**) and $\textbf{V}$ (**n x n**) such that matrix $\textbf{A}$ can be decomposed as follows, using the SVD:

$$\textbf{A} = \textbf{U} \Sigma V^{T}$$

where $\Sigma$ is an **m x n** diagonal matrix having form

$$\Sigma = 
\begin{bmatrix}
    \sigma_{1}      & 0 & 0 & \dots & 0 & 0 \\
    0      & \sigma_{2} & 0 & \dots & 0 & 0 \\
    0      & 0 & \sigma_{3} & \dots & 0 & 0 \\
    ...       & ... & ... & ... \\
    0      & 0 & 0 & \dots & \sigma_{p} & 0 \\
\end{bmatrix}$$ 

where $\sigma_{1} \geq \sigma_{2} \geq \dots \geq \sigma_{p} \geq 0$, $p = \min(m,n)$, and ${\sigma_{i}}$ are the singular values of A. The columns of $U$ are the left singular vectors, and the columns of $V$ are the right singular vectors. If matrix A is rank deficient, then one or more of the singular values will be zero.

Using the SVD, the pseudo-inverse of a matrix can be easily computed. Let A be decomposed as $\textbf{A} = \textbf{U} \Sigma V^{T}$. Then:

$$A^{+} = \textbf{V} \Sigma^{+} U^{T}$$

where \Sigma^{+} takes the form

$$\Sigma^{+} = 
\begin{bmatrix}
    1/\sigma_{1}      & 0 & 0 & \dots & 0 & 0 \\
    0      & 1/\sigma_{2} & 0 & \dots & 0 & 0 \\
    0      & 0 & 1/\sigma_{3} & \dots & 0 & 0 \\
    ...       & ... & ... & ... \\
    0      & 0 & 0 & \dots & 1/\sigma_{p} & 0 \\
\end{bmatrix}$$ 

##Rank Deficient Least Squares using SVD

Let $\textbf{A}$ be an **m x n** real matrix with SVD $\textbf{A} = \textbf{U} \Sigma V^{T}$. Let $\textbf{b}$ be an **m x 1** vector. Assume k = rank($\textbf{A}$).

Assuming $\Sigma_{1}$ is the leading submatrix of $\Sigma$ corresponding to k positive singular values, the compact SVD of $\textbf{A}$ can be written as:

$$\textbf{A} = U_{1} \Sigma V_{1}^{T}$$

where $U_{1} = [u_{1}, u_{2}, \dots , u_{k}]$, $V_{1} = [v_{1}, v_{2}, \dots , v_{k}]$, and

$$\Sigma_{1} = 
\begin{bmatrix}
    \sigma_{1}      & 0 & 0 & \dots & 0 \\
    0      & \sigma_{2} & 0 & \dots & 0 \\
    0      & 0 & \sigma_{3} & \dots & 0 \\
    ...       & ... & ... & ... \\
    0      & 0 & 0 & \dots & \sigma_{k} \\
\end{bmatrix}$$ 

where $\sigma_{1} \geq \sigma_{2} \geq \dots \geq \sigma_{k} \geq 0$.

The *minimum-norm solution* to the linear least squares problem is given by:

$$\hat{x} = A^{+} \textbf{b} = V_{1} \Sigma_{1}^{-1} U_{1}^{T} \textbf{b}$$

where $A^{+} = V_{1} \Sigma_{1}^{-1} U_{1}^{T}$ is the pseudo-inverse of $\textbf{A}$.

So, given a situation where you are attempting to solve a linear least squares problem with a rank deficient, non-invertible matrix, the system can be solved by determining the SVD of the matrix, and then using this SVD to compute the pseudo-inverse needed for the least squares solution.

##Testing
* Let's test out the items discussed above on some data.
* I'll use the Bark transform and cosine matrix from the frequency domain least squares filter design as test data.

####Add the BarkClass folder to the system's path, import the Bark class, and instantiate a Bark object.

```python
import sys
sys.path.append('BarkClass')
```

```python
from Bark import Bark
```

```python
nfft=2048
fs=48000
nfilts=109
version="rasta"
B_trans = Bark(nfft, fs, nfilts, version)

# Get the Bark transform matrix from the Bark object
B = B_trans.W[:,0:B_trans.nfft/2]
```

####Construct the cosine matrix

```python
L = nfft/2
M = 64  # Number of filter coefficients

w = (pi/L)*(np.arange(L)).T  # Frequency grid

# Rectangular C matrix
w_rs = w.reshape(w.shape[0],1)
M_rs = np.arange(M)
M_rs = M_rs.reshape(M_rs.shape[0],1)
C = cos(np.dot(w_rs,M_rs.T))
```

```python
print 'Bark transform shape: ', B.shape
print 'Cosine matrix shape: ', C.shape
```

####The Bark domain filter design using Least Squares depends on the matrix BC (matrix product of Bark transform B and cosine matrix C). Construct this matrix.

```python
# Matrix product of Bark transform (B) and cosine matrix (C)
BC = np.dot(B,C)

print BC.shape
```

####Matrix BC is not square, so it needs to be full rank in order for us to be able to compute its pseudo-inverse. Let's check its rank.

```python
BC_rank = np.linalg.matrix_rank(BC)

print 'Shape of BC: ', BC.shape
print 'Rank of BC: ', BC_rank
```

Letting k = rank(BC) and BC have dimension m x n, we see that k < m and k < n. So BC is not full rank here, and thus is rank deficient.

####What about matrix C? Is it full rank?

```python
C_rank = np.linalg.matrix_rank(C)

print 'Shape of C: ', C.shape
print 'Rank of C: ', C_rank
```

Here, we see that, for matrix C of dimension m x n, rank(C) = n, so C is full rank (full column rank). So its pseudo-inverse should exist and be able to be computed using the method given above.

####Let's compute the pseudo-inverses of BC and C using the method above, just to verify that the rank deficiency produces the result that we expect.

```python
C_inv = np.dot(np.linalg.inv(np.dot(C.T,C)),C.T)
BC_inv = np.dot(np.linalg.inv(np.dot(BC.T,BC)),BC.T)

# Compute the matrix product of the inverses and the original matrices to see
# how close they are to the identity matrix.
I_C_approx = np.dot(C_inv,C)
I_BC_approx = np.dot(BC_inv,BC)

fig = plt.figure(0, (16,8))

ax = fig.add_subplot(121)
ax.imshow(I_C_approx)
ax.set_aspect('auto')
ax = fig.add_subplot(122)
ax.imshow(I_BC_approx)
ax.set_aspect('auto')
```

The plots above show that, as expected, the pseudo-inverse process produces a valid inverse for matrix C but fails for matrix BC.

Because of this, we will need to explore the SVD approach for computing its pseudo-inverse, and for ultimately being able to solve the least squares problem. Luckily, there is a Numpy function for computing the pseudo-inverse using the SVD approach. Let's explore this function a bit.

####Compute the SVD and the pseudo-inverse of matrix BC using the SVD approach in Numpy

```python
U_BC, S_BC, V_T_BC = np.linalg.svd(BC)

BC_pinv = np.linalg.pinv(BC)
```

```python
# Plot the singular values of BC (decreasing order)
plot(S_BC)
```

```python
# Compute the matrix product of the inverse and the original matrix to see
# how close they are to the identity matrix.
I_BC_papprox = np.dot(BC_pinv, BC)

fig = plt.figure(0, (8,8))

ax = fig.add_subplot(111)
ax.imshow(I_BC_papprox)
ax.set_aspect('auto')
```

Not bad! This is much closer to the identity matrix than what was being produced before using the non-SVD pseudo-inverse approach. But how close is it? One thing we could look at is the square error between the approximate identity matrices and the actual identity matrix.

####Compute the square error between the approximate identity matrices and the actual identity matrix.

```python
diff_C = I_C_approx - np.eye(I_C_approx.shape[0])
diff_BC = I_BC_papprox - np.eye(I_BC_papprox.shape[0])

sq_error_C = np.mean(np.power(diff_C,2))
sq_error_BC = np.mean(np.power(diff_BC,2))

print 'Square Error (C): ', sq_error_C
print 'Square Error (BC): ', sq_error_BC
```

From the square error values, it is clear that the nice full-rank characteristic of matrix C makes for a very close approximation of the identity matrix when taking the product of the matrix and its inverse. 

The square error from BC is much larger, but still a very small value! This appears to be working the way we would like for it to.
